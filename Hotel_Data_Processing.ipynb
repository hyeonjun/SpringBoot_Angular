```python
import pandas as pd
import numpy as np
file_path = '/content/drive/MyDrive/data.xlsx'
df = pd.read_excel(file_path)
df.head(1)
```

```python
df.isnull().sum()
```

```python
df['상세정보'].replace(' ', np.nan, inplace=True)
df.dropna(axis=0, subset=['상세정보'], inplace=True)
for c in df.columns:
    if df[c].isnull().sum() > 350:
        df.drop(c, axis=1, inplace=True)
df.drop('규모', axis=1, inplace=True)
df.drop('객실 수', axis=1, inplace=True)
df.drop('예약 안내', axis=1, inplace=True)
```

```python
df.index=list(i for i in range(df['명칭'].count()))
```

```python
for i in range(df['명칭'].count()):
    if '[' in df.loc[i, '명칭']:
        name = df.loc[i, '명칭']
        df.loc[i, '명칭'] = name[:name.index('[')]
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df_hotel = df.iloc[:, :13]
columns = ['hotelName', 'postalCode', 'phone', 'address', 'lat', 'long', 'summary', 'roomType', 'parking', 'cooking', 'checkIn', 'checkOut', 'reservationPage']
for x, y in zip(df_hotel.columns, columns):
    df_hotel.rename(columns={x:y}, inplace=True)
df_hotel.head(3)
```

```python
df_hotel_category = df.iloc[:, 15:-1]
df_hotel_category.head(1)
```

```python
df_hotel_category.insert(0, 'hotelName', df.iloc[:, 0])
df_hotel_category.head(1)
```

```python
category_lst = list(df_hotel_category.columns[1:])
for c in category_lst:
    df_hotel_category[c].replace('없음', False, inplace=True)
    df_hotel_category[c].replace('있음', True, inplace=True)
df_hotel_category.head(10)
```

```python
df_hotel_category.drop('캠프화이어', axis=1, inplace=True)
```

```python
columns = ['seminar', 'sports', 'sauna', 'beauty', 'karaoke', 'BBQ', 'bicycle', 'fitness', 'publicPC', 'publicShower']
for x, y in zip(df_hotel_category.columns[1:], columns):
    df_hotel_category.rename(columns={x:y}, inplace=True)
df_hotel_category.head(3)
```

```python
df_hotel_detail = pd.DataFrame({
    'hotelName':df.iloc[:, 0],
    'detail':df.iloc[:, -1]
})
df_hotel_detail.head(10)
```
